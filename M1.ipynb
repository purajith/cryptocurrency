```python
data1 = data.drop(["Unnamed: 0","Date"], axis = 1)
```

```python
for i,j in enumerate(data["Crypto"].unique()):
    data1["Crypto"] = data1["Crypto"].replace({j: i})
    print(j , " is renamed as :", i)
    
```

correlation of all Crypto and  indidual Crypto

```python
#convert int to floats
data1["Crypto"] = data1["Crypto"].astype(str).astype(float)
data1.info()
```

```python
# Individual correlation for all aCrypto Currency
for i, j  in enumerate(data1['Crypto'].unique()):
    df = data1[data1["Crypto"] == i]
    print(df.corr(),"\n")
```
